### Sobre o dataset:

Os dados referem-se ao consumo de combustível  em milhas por galão (mpg):<br/> 

O período entre 1970 e 1982 marcou uma mudança significativa na indústria automobilística dos Estados Unidos. A produção americana mudou de carros pesados e potentes de seis e oito cilindros com baixo consumo de combustível para carros mais leves e menos potentes de quatro cilindros com maior eficiência de combustível. A indústria automotiva global - incluindo americanos e seus concorrentes europeus e japoneses - aumentou o total de milhas por galão (MPG) ao se concentrar em carros de quatro cilindros e torná-los mais eficientes em termos de combustível.

#### Agrupando por origem = usa

```python
grupo_usa = consumo.groupby(['origin']).get_group('usa')
grupo_usa
```

#### Tratando dados duplicados

```python
#Removendo nomes duplicados
consumo.drop_duplicates()

# Não há dados duplicados 
```

```python
#Função pra primeira letra maiuscula
def trata(entrada):
    texto_max = str.capitalize(entrada)
    return texto_max
```

```python
#Aplicando a função na coluna ORIGIN
consumo['origin'] = consumo['origin'].apply(trata)
```

```python
#Criando coluna Km/l
consumo['Consumo Km/L'] = consumo['mpg'] * 0.43
```

```python
consumo
```

#### Usando a função pivot_table()

```python
# Agrupando dados por ano do modelo, cilindros e média de mpg

media = pd.pivot_table(consumo[['model_year','mpg','cylinders']], index = ['model_year','cylinders'])
media
```

```python
# Agrupando dados por origem, cilindros e média de cavalos de potência

grupo2 = pd.pivot_table(consumo[['origin','cylinders','horsepower']],index = ['origin','cylinders'])
grupo2
```

#### Criando coluna continente

```python
# Função para verificar a qual continente cada carro pertence e colocando a primeira letra maiúscula

def continente(entrada):
    for linha in (consumo['origin']):
        if (entrada == 'Usa'):
            cont = 'américa'
        elif(entrada == 'Japan'):
                cont = 'asia'
        else:
            cont = 'europa'
        
        cont_max = str.capitalize(cont)
        return cont_max
```

```python
# Aplicando a função na coluna continente

consumo['Continente'] = consumo['origin'].apply(continente)
```

```python
consumo.head(20)
```

####  Criando nova coluna de relação peso/potência

```python
consumo['Peso/Potência'] = (consumo['weight'] / consumo['horsepower'])
```

```python
consumo
```

```python
# Gráfico de disperção da relação Km/l / Peso/potência

plt.figure(figsize = (10,5))
plt.scatter(x = consumo['Peso/Potência'], y = consumo['Consumo Km/L'], alpha = 0.5)
plt.title('Gráfico de Disperção de Peso por Consumo', fontsize = 15)
plt.xlabel('Peso/Potência')
plt.ylabel('Km/L')

plt.show()
```
